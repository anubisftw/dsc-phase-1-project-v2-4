```python
def sep(genres):
    sep = fulldf[genres].str.cat(sep = '|')
    
    # Giving pandas series and storing them seperately
    data = pd.Series(sep.split('|'))
    
    # Count them and arraging in descending order
    count = data.value_counts(ascending= False)
    
    return count
```
